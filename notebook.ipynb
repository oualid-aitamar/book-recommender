```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("books.csv", on_bad_lines='skip', engine='python')
```

```python
df.columns = df.columns.str.strip()
```

```python
df.replace("", pd.NA, inplace=True)
df.isnull().sum()
```

```python
df["authors"] = df["authors"].apply(lambda x: x.split("/")[0])
```

```python
df["average_rating"] = pd.to_numeric(df["average_rating"], errors="coerce")
```

```python
df["isbn13"] = df["isbn13"].astype(str)
df["isbn"] = df["isbn"].astype(str)
```

```python
df["publication_date"] = pd.to_datetime(df["publication_date"], errors="coerce")
```

```python
df["year"] = df["publication_date"].dt.year
```

```python
df = df.drop_duplicates(subset=["title", "authors"])
```

```python
df["title"] = df["title"].str.strip()
df["title"] = df["title"].str.replace(r"[^a-zA-Z0-9\s]", "", regex=True)
```

```python
df.to_csv("books_clean.csv", index=False)
```
